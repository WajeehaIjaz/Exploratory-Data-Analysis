# Relationship Analysis

```python
corelation=Student.select_dtypes(include=np.number).corr()
```

Heatmap visualizes correlation between variables using color intensity for quick pattern detection

```python
sns.heatmap(corelation,xticklabels=corelation.columns,yticklabels=corelation.columns,annot=True)
```

Pairplot shows pairwise relationships and distributions between all variables in the dataset

```python
sns.pairplot(Student)
```

Relplot shows relationship between Attendance and CGPA, colored by Grade for comparison

```python
sns.relplot(x='Attendance',y='cgpa',hue='Grade',data=Student)
```

Distplot shows the distribution of CGPA values in the dataset

```python
sns.distplot(Student['cgpa'])
```

Catplot (boxplot) shows the distribution of CGPA and highlights outliers

```python
sns.catplot(x='cgpa',data=Student,kind='box')
```
